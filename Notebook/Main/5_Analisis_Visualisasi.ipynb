```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker

buzz_df['created_at'] = pd.to_datetime(buzz_df['created_at'])

# Extract the date part (removing time)
buzz_df['created_at'] = buzz_df['created_at'].dt.date

# Group by date and sum the 'appearance_count'
daily_counts = buzz_df.groupby('created_at')['appearance_count'].sum().reset_index()

# Plot using Seaborn
sns.set(style="whitegrid", context="talk")
plt.figure(figsize=(14, 7))
sns.lineplot(data=daily_counts, x='created_at', y='appearance_count', marker='o')

def custom_formatter(x, pos):
    if x >= 100000:
        return f'{int(x/100000)}00 K'
    else:
        return str(int(x))

# Add titles and labels
plt.title('Pergerakan Jumlah Buzzer Berdasarkan Waktu', fontsize=16, fontweight='bold')
plt.xlabel('Tanggal (2024)', fontsize=14)
plt.ylabel('Kemunculan Buzzer', fontsize=14)
plt.xticks(rotation=45)

plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%m-%d'))
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(custom_formatter))

# Show plot
plt.tight_layout()
plt.show()
```
