```python
import re

def split_camel_case(string):
  words = re.findall(r'[A-Z]?[a-z]+|[A-Z]+(?=[A-Z]|$)', string)
  return ' '.join(words)
```

### Preprocess

```python
# handle casefold, mentions, links, html char, and whitespaces
result_df = pd.read_pickle('/content/drive/MyDrive/Satria Data pt 2/Semifinal/Dataset/dav/result_df_clean.pkl')
df = result_df[['username', 'tcode', 'content']]
```

```python
import json

# Load slang word dictionary from JSON file
with open('/content/drive/MyDrive/Satria Data pt 2/Semifinal/Dataset/LDA/combined_slang_words.txt', 'r') as file:
    slang_dict = json.load(file)
```

```python
with open('/content/drive/MyDrive/Satria Data pt 2/Semifinal/Dataset/LDA/combined_stop_words.txt', 'r') as file:
    stopwords = [line.strip() for line in file]
```

```python
import pandas as pd
import re

# PERLU HANDLE UNICODE
def preprocess_tweet(tweet):
    pattern = r'[\W]+'
    cleaned_text = re.sub(pattern, ' ', tweet)
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text).strip()

    words = cleaned_text.split()
    replaced_words = [slang_dict.get(word, word) for word in words]
    clean_slang_tweet = ' '.join(replaced_words)

    words = clean_slang_tweet.split()
    filtered_words = [word for word in words if word not in stopwords]
    clean_slang_stopword_tweet = ' '.join(filtered_words)

    return clean_slang_stopword_tweet

df['content'] = df['content'].apply(preprocess_tweet)
df
```
