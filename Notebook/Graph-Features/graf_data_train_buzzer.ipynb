# Create Graf

```python
import time
import networkx as nx
```

```python
df_train.columns
```

```python
df = df_train[['label']]
df.columns = ['class']
df.shape
```

```python
start = time.time()

texts = df_train['content'].tolist()

G = nx.Graph()
for idx, row in df.iterrows():
  G.add_node(idx, **row.to_dict())

print(f"Lama proses: {time.time() - start}")
```

```python
nx.write_gexf(G, folder_path + 'Dataset_Part2/surya/graf_train_buzzer.gexf')

print(f"Jumlah node: {G.number_of_nodes()}")
print(f"Jumlah edge: {G.number_of_edges()}")
```

```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
from tqdm import tqdm
import pickle
```

```python
list_tuple_edges = []
threshold = 50

for i in tqdm(range(len(df)), desc="Processing texts"):
  for j in range(i + 1, len(df)):
    similarity = fuzz.ratio(texts[i], texts[j])
    if similarity > threshold:
      list_tuple_edges.append((str(i), str(j)))
```

```python
with open(folder_path + 'Dataset_Part2/surya/list_tuple_graf_train.pkl', 'wb') as f:
  pickle.dump(list_tuple_edges, f)
```
