```python
df = pd.read_pickle(folder_path + '/Dataset/dav/df_clean_base.pkl')
df
```

```python
target_content = df['content'][df['username'] == '@m9DuXzGHz9vgg5cfyPk8sX8UcfN9tmZ5mATo+gANHNQ='].values
target_content = str(target_content[0])
print(target_content)
```

```python
df = df[df['content'] == target_content]
print(len(df))
```

```python
import matplotlib.pyplot as plt

df['date'] = df['created_at'].dt.date
tweet_counts = df.groupby('date').size()

# Plot the histogram
plt.figure(figsize=(10, 6))
tweet_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Date')
plt.ylabel('Number of Tweets')
plt.title('Distribution of Tweets per Date')
plt.xticks(rotation=45)
plt.show()
```
