```python
import time
import pandas as pd

from fuzzywuzzy import fuzz
from fuzzywuzzy import process

import networkx as nx
from tqdm import tqdm
import random
```

```python
import os
import multiprocessing as mp

# Get the number of CPU cores
num_cores = os.cpu_count()
print(f"Number of CPU cores available: {num_cores}")
```

```python
df.shape
```

```python
df['date'] = df['created_at'].dt.day
df['date'] = df['date'].astype(int)

df = df[['content', 'appearance_count', 'date']]
df.dtypes
```

```python
texts = df['content'].tolist()
```

```python
# Your function to process a range of the loop
def process_range(start, end, df, texts, threshold, result_queue):
    list_tuple_edges = []

    for i in tqdm(range(start, end), desc=f"Processing texts {start}-{end}"):
        current_date = df['date'].iloc[i]

        temp_df = df[(df['date'] >= current_date - 2) & (df['date'] <= current_date + 2)]
        temp_df = temp_df.sample(n=5000, random_state=random.randint(1, 20))

        for j in temp_df.index:
            similarity = fuzz.ratio(texts[i], texts[j])
            if similarity > threshold:
                list_tuple_edges.append((str(i), str(j)))

    # Use a queue to return the result
    result_queue.put(list_tuple_edges)
```

```python
if __name__ == '__main__':
    start = time.time()
    # Define ranges for multiprocessing
    base_line = 2_000_000
    up_line = 2_100_000

    num_processes = num_cores
    range_size = (up_line - base_line) // num_processes
    ranges = [(base_line + i * range_size, base_line + (i + 1) * range_size) for i in range(num_processes)]

    # Create a multiprocessing queue to collect results
    result_queue = mp.Queue()

    # Create and start processes
    processes = []
    for start, end in ranges:
        p = mp.Process(target=process_range, args=(start, end, df, texts, 80, result_queue))
        p.start()
        processes.append(p)

    # Collect results from all processes
    all_edges = []
    for p in processes:
        all_edges.extend(result_queue.get())
        p.join()

    # Now you can use all_edges to add to the graph
    print("Number of edges:", len(all_edges))
    print(f"Lama proses: {time.time() - start}")
```

```python
import pickle

with open(folder_path + 'Dataset_Part2/surya/tuple_2_000_000_sd_2_100_000.pkl', 'wb') as f:
  pickle.dump(all_edges, f)
```
