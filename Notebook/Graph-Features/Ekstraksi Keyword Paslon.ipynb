# Buat DataFrame masing2 paslon

```python
pattern = '|'.join(paslon1_keyterms)

paslon1_df = hashtags_df[hashtags_df['hashtag'].str.lower().str.contains(pattern)]
paslon1_df.sort_values(by='count', ascending=False, ignore_index=True, inplace=True)
paslon1_df.to_excel(folder_path + 'Dataset_Part2/keyterms_paslon/hashtag_paslon1.xlsx')
paslon1_df
```

```python
pattern = '|'.join(paslon2_keyterms)

paslon2_df = hashtags_df[hashtags_df['hashtag'].str.lower().str.contains(pattern)]
paslon2_df.sort_values(by='count', ascending=False, ignore_index=True, inplace=True)
paslon2_df.to_excel(folder_path + 'Dataset_Part2/keyterms_paslon/hashtag_paslon2.xlsx')
paslon2_df
```

```python
pattern = '|'.join(paslon3_keyterms)

paslon3_df = hashtags_df[hashtags_df['hashtag'].str.lower().str.contains(pattern)]
paslon3_df.sort_values(by='count', ascending=False, ignore_index=True, inplace=True)
paslon3_df.to_excel(folder_path + 'Dataset_Part2/keyterms_paslon/hashtag_paslon3.xlsx')
paslon3_df
```
