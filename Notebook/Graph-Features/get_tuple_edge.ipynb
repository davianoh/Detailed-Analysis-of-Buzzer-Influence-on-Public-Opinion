```python
import time
import pandas as pd

from fuzzywuzzy import fuzz
from fuzzywuzzy import process

import networkx as nx
from tqdm import tqdm
import random
```

```python
df['date'] = df['created_at'].dt.day
df['date'] = df['date'].astype(int)

df = df[['content', 'appearance_count', 'date']]
df.dtypes
```

```python
texts = df['content'].tolist()
```

```python
list_tuple_edges = []
threshold = 80

for i in tqdm(range(1_950_000, 2_000_000), desc="Processing texts"):
    current_date = df['date'].iloc[i]

    temp_df = df[(df['date'] >= current_date - 2) & (df['date'] <= current_date + 2)]
    temp_df = temp_df.sample(n=5000, random_state=random.randint(1, 20))

    for j in temp_df.index:
        # Check if both i and j are existing nodes in the graph
        similarity = fuzz.ratio(texts[i], texts[j])
        if similarity > threshold:
          list_tuple_edges.append((str(i), str(j)))
```

```python
import pickle

with open(folder_path + 'Dataset_Part2/surya/tuple_1_950_000_sd_2_000_000.pkl', 'wb') as f:
  pickle.dump(list_tuple_edges, f)
```
